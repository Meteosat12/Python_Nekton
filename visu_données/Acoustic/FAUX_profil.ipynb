```python
import numpy as np
import pandas as pd
import xarray as xr
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import sys
```

```python
df_sv_final = pd.read_csv("/data/rd_exchange/asauvebois/df_svMean_Plot.csv")
```

## Déjà un graphe par province et après on verra

```python
selected_prov= 101.0
```

```python
fig, ax = plt.subplots(figsize=(6, 8))
selected_prov= 101.0
sub_prov = df_sv_final[df_sv_final["province"]== selected_prov]

for mois in  sorted(sub_prov["mois"].unique()):
    sub_mois = sub_prov[sub_prov["mois"] == mois] 
    ax.plot(sub_mois["sv_mean_db"], sub_mois["profondeur"], ms=3,  label=f"Mois {int(mois)}")


ax.invert_yaxis()
ax.set_xlabel("Sv moyen (dB)")
ax.set_ylabel("Profondeur (m)")
ax.set_title(f"Climatologie de profils verticaux mensuels de sv pour la province {selected_prov}")
ax.grid(alpha=0.3)
ax.legend(fontsize=7, ncol=2)
plt.tight_layout()
plt.show()
```

## Avec plusieurs graphes ?

```python
plt.figure(figsize=(22, 16))

subplot_idx = 1


for sub_prov in df_sv_final["province"].unique(): 
    
    plt.subplot(5, 2, subplot_idx)
    
    for mois in  sorted(sub_prov["mois"].unique()):
        sub_mois = sub_prov[sub_prov["mois"] == mois] 
        ax.plot(sub_mois["sv_mean_db"], sub_mois["profondeur"], ms=3,  label=f"Mois {int(mois)}")


    ax.invert_yaxis()
    ax.set_xlabel("Sv moyen (dB)")
    ax.set_ylabel("Profondeur (m)")
    ax.set_title(f"Profils verticaux mensuels de sv pour la province {selected_prov}")
    ax.grid(alpha=0.3)
    ax.legend(fontsize=7, ncol=2)

    subplot_idx +=1


plt.suptitle(f"Climatologie mensuelle du mnkc sur des profils verticaux sv", fontsize=14)
plt.tight_layout()
plt.show()
```
